## Data Visualization with Matplotlib

##### Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics.

```python
# install matplotlib
! pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
```

```python
#Simple plot
plt.plot(x,y)
```

```python
# Create a line plot
plt.plot(x,y)
#add title
plt.title("Basic Line Plot")
#Labling X and Y axis
plt.xlabel("X")
plt.ylabel("Y")
```

```python
# Customize the line plot
plt.plot(x,y,color="red")
```

```python
plt.plot(x,y,color="red",linestyle="-")
```

```python
plt.plot(x,y,color="red",linestyle="--")
```

```python
plt.plot(x,y,color="red",linestyle="-.")
```

```python
# marker
plt.plot(x,y,color="red",linestyle="--",marker="o")
```

```python
plt.plot(x,y,color="red",linestyle="--",marker="*")
```

```python
# add line width
plt.plot(x,y,color="red",linestyle="--",marker="o",linewidth=2)
```

```python
# add markersize
plt.plot(x,y,color="red",linestyle="--",marker="o",linewidth=2,markersize=9)
```

```python
#add grid
plt.plot(x,y,color="red",linestyle="--",marker="o",linewidth=2,markersize=9)
plt.grid()
# plt.show()
```

```python
## Multiple Plots
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]
```

```python
plt.figure(figsize=(9,5))
```

```python
plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x,y1,color="green")
plt.title("Plot 1")
# plt.subplot is used to create multiple plots
plt.subplot(2,2,2)
plt.plot(x,y2,color="red")
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(y1,x,color="blue")
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(y2,x,color="yellow")
plt.title("Plot 4")
```

```python
plt.figure(figsize=(9,5))
plt.subplot(4,1,1)
plt.plot(x,y1,color="green")
plt.title("Plot 1")
# plt.subplot is used to create multiple plots
plt.subplot(4,1,2)
plt.plot(x,y2,color="red")
plt.title("Plot 2")

plt.subplot(4,1,3)
plt.plot(y1,x,color="blue")
plt.title("Plot 3")

plt.subplot(4,1,4)
plt.plot(y2,x,color="yellow")
plt.title("Plot 4")
```

```python
# Create a Scatter Plot
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.scatter(x,y,color="green",marker="*")
```

```python
# Pie chart

values = [1,2,3,4,5]

plt.pie(values)
```

```python
labels = ['A','B','C','D']
sizes = [30,20,40,10]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.2,0,0,0) ##move out the 1st slice
plt.pie(sizes,labels=labels,colors=colors)
```

```python
# Explode is used to show the 1st slice
plt.pie(sizes,explode=explode,labels=labels,colors=colors,)
```

```python
#autopct stands for auto percentage
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

## Data Visualization with Seaborn

#### Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.Seaborn helps in creating complex visualizations with just a few lines of code.

```python
#install seaborn
! pip install seaborn
```

```python
import seaborn as sns
```

```python
# Basic Plotting With Seaborn
tips = sns.load_dataset('tips')
tips
```

```python
# Create a scatter plot
plt.scatter(tips['total_bill'],tips['tip'])
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
```

```python
# Line Plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
x = tips['size']
y = tips['total_bill']
```

```python
plt.plot(x,y)
```

```python
# Categorical Plot
# Bar Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar Plot of Total bill by size")
plt.show()
```

```python
tips.info()
```

```python
sns.barplot(x='size',y='smoker',data=tips, )
plt.title("Bar Plot of Total bill by size")
plt.show()
```

```python
# Box plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box Plot of Total bill by day")
plt.show()
```

```python
#box plot in matplotlib
plt.boxplot(tips['total_bill'],labels=['Total Bill'])
plt.title("Box Plot of Total bill")
plt.show()
```

```python
# Voilin Plot
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Violin Plot of Total bill by day")
plt.show()
```

```python
sns.violinplot(x= 'size', y = 'total_bill', data=tips)
plt.title("Violin Plot of Total bill by size")
plt.show()
```

```python
sns.boxplot(x= 'size', y = 'total_bill', data=tips)
plt.title("Violin Plot of Total bill by size")
plt.show()
```

```python
#Histogram
# kde = kernel density estimation
sns.histplot(tips['total_bill'],bins=10,)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
# KDE Plot
sns.kdeplot(tips['size'])
```

```python
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# Pairplot
# It is used to see the relationship between different variables
# Can take only numerical variables
sns.pairplot(tips)
plt.show()
```

```python
# Heatmap
corr = tips.corr()
corr
```

```python
corr = tips[['total_bill','tip','size']].corr()
corr
```

#### this means if the total bill increases by one unit then tip also increases by 0.675734 unit and size also increases by 0.598315

```python
sns.heatmap(corr)
```

```python
# annotation is true
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
