# Pandas-DataFrame And Series 
1) pandas is a python package that provides high-performance, easy-to-use data structures and data analysis tools 
2) A Series is a one-dimensional array holding data of any type. It can be created using the pd.Series() function
3) A DataFrame is a two-dimensional data structure, it can be created using the pd.DataFrame() function
4) A Datafram can have multiple indexes and heterogeneous columns

```python
import pandas as pd
```

```python
## Series
# Is a 1-D array holding data of any type. It is similar to a column in a table.

data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
## Create a series from dictionary

data = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
series_dict = pd.Series(data)
print(series_dict)   
print(type(series_dict))
```

```python
data = [10,20,30,40,50]
index = ['x', 'y', 'z', 'a', 'b']
series = pd.Series(data, index)
print(series)
```

```python
## DataFrame
## create a DF from a dictionary of lists

data = {
    'Name': ['Tom', 'Jack', 'Steve', 'Ricky'],
    'Age': [28, 34, 29, 42],
    'Rating': [4.23, 3.24, 3.98, 2.56],

}

df = pd.DataFrame(data,index=['rank1','rank2','rank3','rank4'])
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```
